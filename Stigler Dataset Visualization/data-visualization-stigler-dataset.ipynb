```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
print("Setup Complete")
```

Dataset credits:
Yam Peleg, “Calculating An Optimal Diet.” Kaggle, 2021, doi: 10.34740/KAGGLE/DSV/2909755.

```python
#Path of the dataset
diet_data_filepath = "../input/calculating-an-optimal-diet/stigler.csv"

#Get the entire dataset in the diet_data dataframe
diet_data = pd.read_csv(diet_data_filepath, index_col = 'commodity')

#Get only those commodities where unit is 1 lb in partial_diet_data
partial_diet_data = diet_data.loc[diet_data['unit']=='1 lb.']
```

```python
#Scatter plot showing calories contained in food items in partial_diet_data vs their prices in cents
plt.figure(figsize=(16,8))
plt.title("Scatter plot showing calories contained in food items in partial_diet_data vs their prices in cents")
sns.regplot(x=partial_diet_data['price_cents'], y=partial_diet_data['calories'])
```

Inference: The regression line shows that foods of higher prices have lower number of calories compared to foods of lower prices.

```python
#Scatter plot showing protein contained in food items in partial_diet_data vs their prices in cents
plt.figure(figsize=(16,8))
plt.title("Scatter plot showing protein contained in food items in partial_diet_data vs their prices in cents")
sns.regplot(x=partial_diet_data['price_cents'], y=partial_diet_data['protein_g'])
```

Inference: Surprisingly, foods of higher prices have lower amounts of protein compared to foods of lower prices. This is contradictory to the general expectations.

```python
#Scatter plot of calcium contained vs protein contained(both measured in g) in food items in partial_diet_data
plt.figure(figsize=(16,8))
plt.title("Scatter plot of calcium contained vs protein contained(both measured in g) in food items in partial_diet_data")
sns.regplot(x=partial_diet_data['protein_g'], y=partial_diet_data['calcium_g'])
```

Inference: There is no strong correlation between the amount of protein and the amount of calcium contained in 1 lb. of the foods in the dataset extracted from the data used for the stingler problem. However, since the regression line has a positive slope, the overall trend is that the amount of calcium increases as the amount of protein(both in grams) increases.

```python
#Histogram of Vitamin C contained(measured in mg) Scatter plot of calcium contained vs protein contained(both measured in g) in food items in partial_diet_data
plt.figure(figsize=(16,8))
plt.title("Histogram of Vitamin C contained(measured in mg)")
sns.distplot(a=partial_diet_data['vitamin_c_mg'], kde=False)
```

Inference: Most of the foods in the partial data set extracted from the data used for the stingler problem have very low amounts(<1000mg per lb.) of Vitamin C.

```python
#Bar chart showing prices of all the food items in partial_diet_data
plt.figure(figsize=(16,8))
plt.xticks(rotation=90)
plt.title("Bar chart showing prices of all the food items in partial_diet_data")
sns.barplot(x=partial_diet_data.index, y=partial_diet_data['price_cents'])
```

Inference: 
Most expensive food - Veal Cutlets, 
Least expensive food - Onions


```python
#Bar chart showing calories contained by different food items in partial_diet_data
plt.figure(figsize=(16,8))
plt.xticks(rotation=90)
plt.title("Bar chart showing calories contained by different food items in partial_diet_data")
sns.barplot(x=partial_diet_data.index, y=partial_diet_data['calories'])
```

Inference:
Food with the most calories - Lard, 
Food with the least calories - Spinach

```python
#Heatmap of contents of Iron, Vitamin B1, Vitamin B2, Niacin(in mg) for every food item in partial_diet_data
plt.figure(figsize=(16,8))
plt.title("Heatmap of contents of Iron, Vitamin B1, Vitamin B2, Niacin(in mg) for every food item in partial_diet_data")
new_partial_diet_data = partial_diet_data.iloc[: , [5, 7, 8, 9]].copy()
sns.heatmap(data=new_partial_diet_data, annot=False)
plt.xlabel("Iron, Vitamin B1, Vitamin B2, Niacin(in mg)")
```
